We redo this [blog post](https://www.algebraicjulia.org/blog/post/2021/09/cset-graphs-4/) except for the co-Heyting algebra structure, namely Lawvere's non-$S$ operation $\sim\!\! S$ and the complements $S \setminus T$:

```julia
LoadPackage( "FunctorCategories" )
```

```julia
FinQuivers
```

```julia
UnderlyingCategory( FinQuivers )
```

```julia
G = CreateQuiver( 8, [ 0,1,  1,2,  2,3,  3,0,  4,5,  6,6,  2,7 ] )
```

```julia
S = Subobject( G, [ 0, 1, 3 ] )
```

```julia
notS = EmbeddingOfPseudoComplementSubobject( S )
```

```julia
notnotS = EmbeddingOfPseudoComplementSubobject( notS )
```

```julia
S_or_notS = EmbeddingOfUnionSubobject( S, notS )
```

```julia
T = Subobject( G, [ 1, 2, 5, 6 ] )
```

```julia
S_and_T = EmbeddingOfIntersectionSubobject( S, T )
```

```julia
S_or_T = EmbeddingOfUnionSubobject( S, T )
```

```julia
S_implies_T = EmbeddingOfRelativePseudoComplementSubobject( S, T )
```
