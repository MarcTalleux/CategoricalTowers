```julia
Q = HomalgFieldOfRationalsInSingular( )
```

```julia
R = Q["x,y"]
```

```julia
K = HomalgMatrix( "[ x, -x ]", 2, 1, R )
```

```julia
J = HomalgMatrix( "[ x, y ]", 2, 1, R )
```

```julia
I = KroneckerMat( K, J )
```

```julia
P = CategoryOfRows( R )
```

```julia
P = LazyCategory( P, primitive_operations = true, show_evaluation = true )
```

```julia
I = I / P
```

```julia
SetLabel( Source( I ), "R^i" )
```

```julia
R = Range( I )
```

```julia
SetLabel( R, "R" )
```

```julia
J = J / P
```

```julia
SetLabel( J, "J" ); SetLabel( Source( J ), "R^j" )
```

```julia
K = K / P
```

```julia
S = SliceCategoryOverTensorUnit( P )
```

```julia
L = PosetOfCategory( S )
```

```julia
I = I / S / L
```

```julia
J = J / S / L
```

```julia
K = K / S / L
```

```julia
IqJ = InternalHom( J, I )
```

```julia
VisualizeInJulia( UnderlyingMorphism( UnderlyingCell( IqJ ) ) )
```

```julia
K == IqJ
```
