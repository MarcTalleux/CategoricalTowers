```julia
G = CreateQuiver( 8, [ 0,1,  1,2,  2,3,  3,0,  4,5,  6,6,  2,7 ] )
```

```julia
S = Subobject( G, [ 0, 1, 3 ] )
```

```julia
notS = EmbeddingOfPseudoComplementSubobject( S )
```

```julia
another_notS = PushoutComplement( UniversalMorphismFromInitialObject( Source( S ) ), S )[2]
```

```julia
notnotS = EmbeddingOfPseudoComplementSubobject( notS )
```

```julia
SandnotS = EmbeddingOfUnionSubobject( S, notS )
```

```julia
T = Subobject( G, [ 1, 2, 5, 6 ] )
```

```julia
SandT = EmbeddingOfIntersectionSubobject( S, T )
```

```julia
SorT = EmbeddingOfUnionSubobject( S, T )
```

```julia
SimpliesT = EmbeddingOfRelativePseudoComplementSubobject( S, T )
```
